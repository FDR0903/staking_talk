```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
def rootfunc(_x, _mudol, _sigmadol, _rho, _a, _b):
    return (2*_x-1)*(1-_x)*(_a*_b-(1-_rho)*_sigmadol**2) + _mudol

def rootfunc(_x, _mudol, _sigmadol, _rho, _a, _b):
    return (2*_x-1)*(1-_x) + (_b * (1-_x) - _mudol)/((1-_rho)*_sigmadol**2)

def rootfunc(_x, _mudol, _sigmadol, _rho, _a, _b):
    return (2*_x-1)*_x + (_b * _x - _mudol)/((1-_rho)*_sigmadol**2)
```

```python
def lol(_thetas, _mu, _mui, _sigma, _rho, _a, _b):
    x = _thetas[0]
    y = _thetas[1]
    #x = min(max(x, 0), 1)
    #y = min(max(y, 0), 1)
    
    a = x - (1 / (1 - _rho**2)) * (
                (_mu * (1 - _rho)
                 - _a
                 + _b * x
                 + _rho * (_a - _b * y)
                 - _rho * _mui / (x + y))
                / (_sigma**2)
            ) 
    b = y - (1 / (1 - _rho**2)) * (
                (_mu * (1 - _rho)
                 + _rho * (_a - _b * x)
                 - _a
                 + _b * y
                 + _mui / (x + y) )
                / (_sigma**2)
            )
    return np.array([a,b ])
    
def lol(_thetas, _mu, _mui, _sigma, _rho, _a, _b):
    x = _thetas[0]
    y = _thetas[1]
    #x = min(max(x, 0), 1)
    #y = min(max(y, 0), 1)
    
    a = x - (1 / (1 - _rho**2)) * (
                (_mu * (1 - _rho)
                 - _a
                 + _b * x
                 + _rho * (_a - _b * y)
                 - _rho * _mui )
                / (_sigma**2)
            ) 
    b = y - (1 / (1 - _rho**2)) * (
                (_mu * (1 - _rho)
                 + _rho * (_a - _b * x)
                 - _a
                 + _b * y
                 + _mui  )
                / (_sigma**2)
            )
    return np.array([a,b ])
```

```python
import numpy as np

from scipy.optimize import minimize

def c(_x,_a,_b):
    return _a - _b*_x
def tomin(_thetas, _mu, _mui, _sigma, _rho, _a, _b):
    x = _thetas[0]
    y = _thetas[1]
    logdrift = x*(_mu-c(x,_a,_b)-(x+y)*_mui)\
                    + y*(_mu - c(y,_a,_b) + x*_mui)\
                    +(1-x-y)*x*_mui\
                    + (-x*x*_sigma*_sigma/2 - y*y*_sigma*_sigma/2 - _rho*x*y*_sigma*_sigma)
    return -logdrift
```

```python
mudol    = .1
mui      = .3
sigmadol = .5
rho      = .4
a        = .01
b        = a

from scipy.optimize import LinearConstraint

linear_constraint = LinearConstraint([[0, 1], [1, 0], [1, 1]], [0, 0, 0], [1, 1, 1])

res = minimize(lambda thetas: tomin(thetas, mudol, mui, sigmadol, rho, a, b),  [0.5,0.5], constraints = [linear_constraint])
#, method='nelder-mead',               options={'xatol': 1e-8, 'disp': True}
res.x
```

```python
mudol    = .1
mui      = .3
sigmadol = .5
rho      = .4
a        = .04
b        = a

muis = np.linspace(0, .9)

thetas = [minimize(lambda thetas: tomin(thetas, mudol, mui, sigmadol, rho, a, b),  [0.5,0.5], constraints = [linear_constraint]).x for mui  in muis]

fig, ax = plt.subplots(1, 1)

def minmiax(x):
    return min(max(x, 0), 1)

ax.plot(muis, [theta[0] for theta in thetas], color='k', lw=3)
ax.plot(muis, [theta[1] for theta in thetas], color='b', lw=3)

ax.xaxis.set_major_formatter(mtick.PercentFormatter(1))
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
ax.legend(['LST','ETH'], handlelength=0.4)
ax.set_xlabel('Issuance tax')
```

```python
# high taw issuance
rhos = np.linspace(0, .9)
thetas = [optimize.root(lambda x: lol(x, mudol, mui, sigmadol, rho, a, b), [1, 1], method='hybr').x
          for rho  in rhos]

fig, ax = plt.subplots(1, 1)

def minmiax(x):
    return min(max(x, 0), 1)
    
ax.plot(rhos, 
         [minmiax(theta[0])/(minmiax(theta[0])+minmiax(theta[1])) for theta in thetas], color='k', lw=3)
ax.plot(rhos, 
         [minmiax(theta[1])/(minmiax(theta[0])+minmiax(theta[1])) for theta in thetas], color='b', lw=3)
ax.xaxis.set_major_formatter(mtick.PercentFormatter(1))
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
ax.legend(['LST','ETH'], handlelength=0.4)
ax.set_xlabel('Correlation (access to DeFi with LSTs)')
```
